```python
#importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Importing the Dataset
df=pd.read_csv('/content/drive/My Drive/datasets/Position_Salaries.csv')
df
```

```python
#Seperating the features and labels
X=df.iloc[:,1:-1].values

y=df.iloc[:,-1].values

print(X, y)
```

Position and the Level of the employee represent the same thing.So in our machine learning model it is sufficient to consider only on feature.In this case we can select the column Level.

```python
#Applying linear regression
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(X ,y)
```

```python
#Visualizing linear regression result

plt.scatter(X , y, color='red')
plt.plot(X,lin_reg.predict(X),color='blue')
plt.title('Truth Or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
#Applying Polynomial Regression with degree 4'

from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=4)
X_poly=poly_reg.fit_transform(X)
lin_reg2=LinearRegression()
lin_reg2.fit(X_poly, y)

```

```python
#Visualizing the results

plt.scatter(X, y, color = 'magenta')
plt.plot(X, lin_reg2.predict(X_poly), color = 'cyan')
plt.title('Polynomial Regression')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
# Predicting the salary for Partner which is 200000 in the original dataset
print(lin_reg.predict([[7]]))

lin_reg2.predict(poly_reg.fit_transform([[7]]))
```

Here, we can observe that polynomial regression  giving pretty accuarate result i.e. 184003 close to 200000 but the linear regression model is a not worth applying on this non linear dataset.
