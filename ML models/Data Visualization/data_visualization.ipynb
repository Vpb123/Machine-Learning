### Matplotlib plots

```
import matplotlib.pyplot as plt
%matplotlib inline
```

```
import numpy as np
x=np.arange(0, 10)
y=np.arange(11, 21)
```

```
plt.scatter(x,y,c='g')
```

```
y=x*x
plt.plot(x,y,'r*', linestyle='dashed',linewidth=2)
```

```
 plt.subplot(3, 3, 1)
 plt.plot(x,y,'r')
 plt.subplot(3, 3, 2)
 plt.plot(x,y,'g')
 plt.subplot(3, 3, 3)
 plt.plot(x,y, )

```

```
np.pi
x=np.arange(0,4*np.pi, 0.1)
y=np.sin(x)
plt.plot(x,y)
```

```
x=[2,8,10]
y=[11,16,9]
x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y)
plt.bar(x2,y2,color='g')
```

```
labels=['python','Ruby', 'C++','java']
explode=(0.4,0,0,0)
sizes=[215,130,210,245]
colors=['green','gold','blue','lightskyblue']
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',shadow=False)
plt.axis('equal')
```

```
import seaborn as sns 
```

### Data Visualization of tips dataset using seaborn

```
df=sns.load_dataset("tips")
```

```
df
```

```
df.dtypes
```

```
df.corr()
```

Visualizing correlations between values

```
sns.heatmap(df.corr())
```

Visualizing the best fit line using for tips to be paid by total_bill paid

```
 sns.jointplot(x='tip',y='total_bill',data=df, kind='reg')
```

Visualizing pairwise relationship using pairplot

```
sns.pairplot(df)
```

```
sns.pairplot(df, hue='sex') 
```

```
sns.displot(df['tip'],kde=True)
```

Plots of categorical data

```
sns.countplot('day',data=df)
```

```
sns.countplot(y='sex', data=df)
```

```
sns.barplot(x='total_bill', y='smoker', data=df)
```

```
sns.boxplot('day' ,'total_bill', data=df, palette='rainbow')
```

```
sns.boxplot(data=df, orient='v')
```

```
sns.boxplot(x="total_bill", y="day", hue="smoker", data=df)
```

```
sns.violinplot(x='total_bill', y='day', data=df, palette='rainbow')
```
