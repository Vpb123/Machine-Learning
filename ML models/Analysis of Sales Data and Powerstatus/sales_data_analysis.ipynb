Task 3 : Find out the per unit price from the data


```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
d=data[["Sales","Quantity","Discount"]]
arr1=round(pd.Series(d["Sales"]),2)
print(arr1)
arr2=pd.Series(d["Quantity"])
arr3=round(pd.Series(d["Discount"]),2)
```

```python
p=(1-arr3)
print(p)
price=round(arr1.divide(p),2)
price_per_unit=round(price.divide(arr2),2)
print(price_per_unit)
```

```python
data['Unit_Price']=price_per_unit
print(data)
```

Task 4 : Find out the monthly revenue and analyze the findings

```python

data['Year'] = data['Order Date'].dt.year
data['Month'] = data['Order Date'].dt.month
print(data)
```

```python
df=data[["Sales","Month"]].groupby("Month")
df1=pd.DataFrame(round(df["Sales"].sum(),2))
df1["Month"]=df1.index
df1.columns
```

```python


plt.bar(df1["Month"],df1["Sales"])
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

Task 5 : Find out the yearly revenue and analyze the findings


```python
df2=pd.DataFrame(round(data[["Sales","Year"]].groupby("Year").sum(),2))
df2.reset_index()
```

```python
df2.plot(use_index=True,legend=True)
plt.ylabel("Revenue")
```

Task 6 : Finding out the monthly growth rate and analyse the findings


```python
d=(df1["Sales"].pct_change())
df1['Growth_Rate']=round(d*100,2)
df1.columns
```

```python
df1
```

```python
plt.plot(df1["Month"],df1["Growth_Rate"])
plt.title("Mothly Growth Rate")
plt.xlabel("Month")
plt.ylabel("Growth Rate")
plt.show()
```

Task 7 : Finding out the most and least sold product id


```python
data_sold=data[["Product ID","Quantity"]].groupby("Product ID").sum()
print(f"The Most Sold Product's ID is {max(data_sold['Quantity'].index)} : {max(data_sold['Quantity'])} Pieces sold.")
print(f"The Least Sold Product's ID is {min(data_sold['Quantity'].index)} : {min(data_sold['Quantity'])} Pieces sold.")
```
