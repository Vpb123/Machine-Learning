```
import numpy as np
import pandas as pd
import requests
import bs4
import lxml.etree as xml
```

```
URL='https://www.macrotrends.net/countries/IND/india/population'
requests.get(URL)
```

```
requests.get(URL,{}).text
```

```
webpage=bs4.BeautifulSoup(requests.get(URL,{}).text,"lxml")
webpage
```

```
webpage.head.title.text
```

```
webpage.body
```

```
tags_element=webpage.find_all(name='table',attrs={"class":"historical_data_table"})[1]
tags_element
```

```

rows = []
for i, row in enumerate(tags_element.tbody.find_all("tr")):
    tds = row.find_all("td")
    
    Year        = tds[0].text
    Population = tds[1].text
    Growth_Rate  = tds[2].text
    
    rows.append((Year,Population,Growth_Rate))
data_frame=pd.DataFrame(rows,columns=['Year','Population','Growth Rate'])
data_frame
```

```
df=data_frame.loc[0:20,'Year':'Growth Rate']
df
```

```
f=df.to_csv('population.csv')
```

```
a=df.to_dict('split')
data=a['data']
```

```
year=[]
pop=[]
growth=[]
for d in data:
  year.append(int(d[0]))
  pop.append(d[1])
  growth.append(float(d[2].replace('%','')))
population=[]
for p in pop:
  a=p.replace(',','')
  population.append(int(a))
  
growth
```

```
import numpy as np
import matplotlib.pyplot as plt
x=np.array(year)
y=np.array(population)
z=np.array(growth)
plt.title('Population of India')
plt.xlabel('Year')
plt.ylabel('Population')
plt.bar(x,y)
```

```
plt.plot(x,z)
plt.ylabel('Growth Rate')
plt.xlabel('Year')
```
