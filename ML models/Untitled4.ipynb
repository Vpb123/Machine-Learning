```
#Create a program which finds out the details of a person from the paragraph given below and save them inside a csv file.
s="Hi my name is SAGAR. I am from PUNE. By profession I am a TRAINER. I have studied ENGINEERING. My favorite sport is CRICKET. My interest area is ML."

l=s.split()
l2=list()
for i in l:
  if i.isupper() and i!="I":
    l2.append(i)
print(l2)
import pandas as pd
df = pd.DataFrame([l2], columns = ['Name', 'City', 'Profession', 'Field of study','fav Sport','Interest Area']) 
print(df)
df.to_csv('data.csv')
```
